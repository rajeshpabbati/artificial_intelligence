# 1. Given an employee id, greet them with their first name.
###	Input
###	E123D
###	Output
###	“Hello, Mr Nitin”

```python
# Defining a function which take input parameter to greet employee
def greet_employee(employee_id):
    # comparing the user input empoyee id with data present in dataframe and storing in a variable
    employee = df[df["EID"] == employee_id]
    # if employee not found return message appropriately
    if employee.empty:
        return "Employee ID not found"
    # extracting employee name from dataset for given eid
    employee_name = employee.iloc[0]["Employee Name"]
    # extracting gender from dataset for given eid
    employee_gender = employee.iloc[0]["Gender"]
    # extracting first name
    first_name = employee_name.split()[0]
    # message and name store in greeting based on gender
    if employee_gender == "M":
        greeting = f"Hello, Mr {first_name}"
    elif employee_gender == "F":
        greeting = f"Hello, Ms {first_name}"
    else:
        greeting = f"Hello, {first_name}"
    # function return the greeting message to calling function
    return greeting
employee_id = input("Enter Employee ID ")
greet_employee(employee_id)
```

# 2. Function for calculating the average salary of each department.

```python
# Defining function to calculate average salary for each department
def averge_salary_by_department():
    # grouping by department and calcualting the mean of the salary
    average_salary = df.groupby("Department")["Salary"].mean()
    # returning the calculated average salary to calling function
    return average_salary
averge_salary_by_department()
```

# 3. Given an Employee id, display respective employees Name, Data of Joining and Department.
### Input
### E123D
### Output
### Nitin Bhole | 01-11-2016 | Products

```python
# defining a function to display employee details which takes eid as parameter
def display_employee_details(employee_id):
    # storing the user entered eid details in variable
    employee = df[df["EID"] == employee_id]
    # return appropriate message if eid not found
    if employee.empty:
        return "Employee ID not found"
    # extract employee name and store in variable
    employee_name = employee.iloc[0]["Employee Name"]
    # extract employee date of joining and store in variable
    employee_doj = employee.iloc[0]["Date of Joining"]
    # extract employee department and store in variable
    employee_dep = employee.iloc[0]["Department"]
    # store employee required details in variable
    greeting = f"{employee_name} | {employee_doj} | {employee_dep}"
    # return the final variable to calling function
    return greeting
employee_id = input("Enter Employee ID")
display_employee_details(employee_id)
```

# 4. Function for sorting employees from Oldest to Newest.

```python
# defining function to sort employees from oldedt to newest
def sorting_oldest_newest():
    # convert date of joining to datetime format, dayfirst is given because it is not in "%m-%d-%y" format and store in
    # result in same column of dataframe
    df["Date of Joining"] = pd.to_datetime(df["Date of Joining"], dayfirst=True)
    # sort by employee date of joining and return to calling function
    return df.sort_values(by="Date of Joining")
sorting_oldest_newest()
```
